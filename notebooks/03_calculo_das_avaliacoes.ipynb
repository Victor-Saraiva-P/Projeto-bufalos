## Resumo rapido

Confere quantas imagens e segmentacoes ficaram com metricas persistidas.


```python
imagens = avaliacao_controller.imagem_repository.list()
imagens_com_ground_truth = sum(
    1
    for imagem in imagens
    if imagem.ground_truth_binarizada is not None
    and imagem.ground_truth_binarizada.area is not None
    and imagem.ground_truth_binarizada.perimetro is not None
)
segmentacoes_avaliadas = sum(
    1
    for imagem in imagens
    for segmentacao in imagem.segmentacoes
    if segmentacao.area is not None
    and segmentacao.perimetro is not None
    and segmentacao.iou is not None
)

print(f"Imagens com ground truth avaliado: {imagens_com_ground_truth}")
print(f"Segmentacoes com metricas persistidas: {segmentacoes_avaliadas}")
print("Para explorar ranking e visualizacoes, siga para o notebook 04.")
```
