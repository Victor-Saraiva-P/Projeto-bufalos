## Imports

imports do projeto

```python
from indice_loader import carregar_indice_excel
from models.avaliacao import Avaliacao
```

## Configuracao

```python
from config import (
    GROUND_OF_TRUTH_OUTPUT,
    MODELOS_PARA_AVALIACAO,
    REMBG_IMAGE_TYPE,
    SEGMENTED_PHOTOS_DIR,
)
```

## Leitura do indice

```python
indice_excel = carregar_indice_excel()
```

## Instancia as avaliacoes

```python
for linha in indice_excel:
    avaliacao = Avaliacao(
        linha.nome_arquivo,
        REMBG_IMAGE_TYPE,
        GROUND_OF_TRUTH_OUTPUT,
        SEGMENTED_PHOTOS_DIR,
        list(MODELOS_PARA_AVALIACAO.keys()),
    )

    print(
        f"A avaliacao {avaliacao.nome_arquivo} teve como area {avaliacao.ground_truth.area} e perimetro {avaliacao.ground_truth.perimetro}"
    )
    for model in avaliacao.segmetacoes:
        print(
            f"Avaliando modelo: {model.modelo}, area: {model.area}, perimetro: {model.perimetro}"
        )
```
