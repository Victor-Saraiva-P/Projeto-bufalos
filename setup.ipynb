Verifica se os arquivos de data estão presentes

```python
from pathlib import Path

data_dir = Path('data')
indice_path = data_dir / 'Indice.xlsx'
mascara_dir = data_dir / 'Mascara'
original_dir = data_dir / 'Original'

def has_images(folder: Path) -> bool:
    if not folder.is_dir():
        return False
    exits = {'.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff', '.gif', '.webp'}
    return any(p.is_file() and p.suffix.lower() in exits for p in folder.iterdir())

errors = []

if not data_dir.is_dir():
    errors.append("Pasta 'data' nao encontrada.")

if not indice_path.is_file():
    errors.append("Arquivo 'Indice.xlsx' nao encontrado em data/.")

if not mascara_dir.is_dir():
    errors.append("Pasta 'Mascara' nao encontrada em data/.")
elif not has_images(mascara_dir):
    errors.append("Pasta 'Mascara' nao tem imagens.")

if not original_dir.is_dir():
    errors.append("Pasta 'Original' nao encontrada em data/.")
elif not has_images(original_dir):
    errors.append("Pasta 'Original' nao tem imagens.")

if errors:
    raise FileNotFoundError('\n'.join(errors))

print('Validacao OK: data/ existe, Indice.xlsx presente, Mascara/ e Original/ com imagens.')
```
